```python
import re
from functools import reduce

class Animal:
    def __init__(self, name, legs, hands, family):
        self.name = name
        self.legs = legs
        self.hands = hands
        self.family = family

    def get_details(self):
        return {
            "Name": self.name,
            "Legs": self.legs,
            "Hands": self.hands,
            "Family": self.family
        }

class Feline(Animal):
    def __init__(self, name):
        super().__init__(name, legs=4, hands=0, family="Cat Family")
        self.characteristics = []

    def get_details(self):
        details = super().get_details()
        details["Characteristics"] = self.characteristics
        return details

class Canine(Animal):
    def __init__(self, name):
        super().__init__(name, legs=4, hands=0, family="Dog Family")
        self.characteristics = []

    def get_details(self):
        details = super().get_details()
        details["Characteristics"] = self.characteristics
        return details

class Tiger(Feline):
    def __init__(self):
        super().__init__(name="Tiger")
        self.characteristics = ["Can Roar", "Lethal Predator"]

class WildCat(Feline):
    def __init__(self):
        super().__init__(name="Wild Cat")
        self.characteristics = ["Can Climb Trees"]

class Wolf(Canine):
    def __init__(self):
        super().__init__(name="Wolf")
        self.characteristics = ["Hunts in Packs", "Has a Leader"]

class Bird:
    def __init__(self, name, legs, wings):
        self.name = name
        self.legs = legs
        self.wings = wings

    def get_details(self):
        return {
            "Name": self.name,
            "Legs": self.legs,
            "Wings": self.wings,
            "Family": "Bird Family"
        }

class FlightBird(Bird):
    def __init__(self, name):
        super().__init__(name, legs=2, wings=2)
        self.characteristics = ["Flies", "Hunts for Food"]

    def get_details(self):
        details = super().get_details()
        details["Characteristics"] = self.characteristics
        return details

class Eagle(FlightBird):
    def __init__(self):
        super().__init__(name="Eagle")
        self.characteristics.append("Flies Extremely High")
        self.characteristics.append("Can See Prey from High Up")

class Zoo:
    def __init__(self):
        self.capacity_animals = 2  # Maximum 2 animals
        self.capacity_birds = 1    # Maximum 1 bird
        self.animals_and_birds = []

    def add_animal_or_bird(self, animal_or_bird):
        # Check if an instance of the same class is already in the zoo
        existing = list(map(lambda a: isinstance(a, type(animal_or_bird)), self.animals_and_birds))

        if any(existing):
            return f"{animal_or_bird.name} (a {type(animal_or_bird).__name__}) is already added to the zoo."

        # Count current animals and birds
        current_animals = len(list(filter(lambda a: isinstance(a, Animal), self.animals_and_birds)))
        current_birds = len(list(filter(lambda a: isinstance(a, Bird), self.animals_and_birds)))

        if isinstance(animal_or_bird, Animal):
            if current_animals < self.capacity_animals:
                self.animals_and_birds.append(animal_or_bird)
                return f"{animal_or_bird.name} added to the zoo."
            else:
                return "Zoo can only have 2 animals!"
        elif isinstance(animal_or_bird, Bird):
            if current_birds < self.capacity_birds:
                self.animals_and_birds.append(animal_or_bird)
                return f"{animal_or_bird.name} added to the zoo."
            else:
                return "Zoo can only have 1 bird!"
        else:
            return "Invalid animal or bird."

    def view_animals_and_birds(self):
        # Create a single string representing the zoo using map and reduce
        details = list(map(lambda a: f"{a.get_details()['Name']} (Family: {a.get_details()['Family']})", self.animals_and_birds))
        return reduce(lambda acc, curr: f"{acc}\n{curr}", details, "Zoo Animals and Birds:")

    def view_canines(self):
        # Use filter function with lambda to find all canines in the zoo
        canines = list(filter(lambda a: isinstance(a, Canine), self.animals_and_birds))
        return [canine.get_details() for canine in canines]

    def view_tigers_regex(self):
        # Use regex to filter out tigers in the zoo
        tiger_pattern = re.compile(r"^Tiger$", re.IGNORECASE)
        tigers = list(filter(lambda a: tiger_pattern.match(a.name), self.animals_and_birds))
        return [tiger.get_details() for tiger in tigers]

# Example usage:
zoo = Zoo()

# Adding animals and birds
print(zoo.add_animal_or_bird(Tiger()))  # Add Tiger
print(zoo.add_animal_or_bird(WildCat()))  # Add Wild Cat
print(zoo.add_animal_or_bird(Eagle()))  # Add Eagle
print(zoo.add_animal_or_bird(Wolf()))  # Attempt to add Wolf - should indicate zoo is full for animals

# Attempt to add another Tiger - should indicate already added
print(zoo.add_animal_or_bird(Tiger()))

# View all animals and birds in the zoo
print(zoo.view_animals_and_birds())

# View canines in the zoo
print("Canines in the zoo:", zoo.view_canines())

# Attempt to view tigers using regex
print("Tigers in the zoo (using regex):", zoo.view_tigers_regex())
```
