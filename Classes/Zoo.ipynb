```python
class Animal:
    def __init__(self, name, legs, hands, characteristics):
        self.name = name
        self.legs = legs
        self.hands = hands
        self.characteristics = characteristics

    def get_info(self):
        return {
            'Name': self.name,
            'Legs': self.legs,
            'Hands': self.hands,
            'Characteristics': self.characteristics
        }

class Feline(Animal):
    def __init__(self, name, characteristics):
        super().__init__(name, legs=4, hands=0, characteristics=characteristics)

class Canine(Animal):
    def __init__(self, name, characteristics):
        super().__init__(name, legs=4, hands=0, characteristics=characteristics)

class Bird:
    def __init__(self, name, legs, wings, characteristics):
        self.name = name
        self.legs = legs
        self.wings = wings
        self.characteristics = characteristics

    def get_info(self):
        return {
            'Name': self.name,
            'Legs': self.legs,
            'Wings': self.wings,
            'Characteristics': self.characteristics
        }

class FlightBird(Bird):
    def __init__(self, name, characteristics):
        super().__init__(name, legs=2, wings=2, characteristics=characteristics)

class Tiger(Feline):
    def __init__(self):
        super().__init__('Tiger', ['Can roar', 'Lethal predator'])

class WildCat(Feline):
    def __init__(self):
        super().__init__('Wild Cat', ['Can climb trees'])

class Wolf(Canine):
    def __init__(self):
        super().__init__('Wolf', ['Hunts in packs', 'Has a leader'])

class Eagle(FlightBird):
    def __init__(self):
        super().__init__('Eagle', ['Can fly extremely high', 'Can see prey from high up in the sky'])

class Zoo:
    def __init__(self):
        self.animals = []  # List to hold animals
        self.bird = None   # To hold the single bird
        self.max_animals = 2
        self.max_birds = 1

    def add_animal(self, animal):
        if len(self.animals) < self.max_animals:
            self.animals.append(animal)
            print(f"Added: {animal.get_info()['Name']}")
        else:
            print("Cannot add more animals! The zoo can hold a maximum of 2 animals.")

    def add_bird(self, bird):
        if self.bird is None:
            self.bird = bird
            print(f"Added: {bird.get_info()['Name']}")
        else:
            print("Cannot add more birds! The zoo can hold only 1 bird.")

    def view_animals_and_birds(self):
        if not self.animals and not self.bird:
            print("The zoo is empty!")
            return

        for animal in self.animals:
            info = animal.get_info()
            print(info)

        if self.bird:
            info = self.bird.get_info()
            print(info)

# Example usage
if __name__ == "__main__":
    zoo = Zoo()

    tiger = Tiger()
    wild_cat = WildCat()
    eagle = Eagle()
    wolf = Wolf()

    # Adding animals and a bird
    zoo.add_animal(tiger)
    zoo.add_animal(wild_cat)
    zoo.add_bird(eagle)

    # Try to add a wolf to demonstrate the capacity limit for animals
    zoo.add_animal(wolf)  # This will not be added since the zoo can hold only 2 animals.

    # Try to add another eagle to demonstrate the capacity limit for birds
    zoo.add_bird(Eagle())  # This will not be added since the zoo can hold only 1 bird.

    # View all animals and birds in the zoo
    zoo.view_animals_and_birds()
```
